```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data1 = map(float, open('prediction/consumption_forcast_test.txt').read().split(','))
data2 = map(float, open('prediction/consumption_test.txt').read().split(','))
```

```python
plt.figure(figsize=(20,10))
plt.title('Consumption Forecasting')
plt.xlabel('hour')
plt.ylabel('value')
plt.plot(data2[:24*30], label='consumption')
plt.plot(data1[:24*30], label='consumption prediction')
plt.legend()
```
